This code determines how age affects insurance costs

```python
# import relevant libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Import insurance.csv into your notebook (Source).

```python
# Load insurance data from insurance.csv
df = pd.read_csv("insurance.csv")
df.head() # check by using head if data was read correctly
```

Use the data in the relevant columns to determine how age affects
insurance costs:

```python
# Extract relevant columns needed to plot scatter plot which are age and charges
age = df['age']
charges = df['charges']
```

Plot a scatter plot with age on the x-axis and charges on the y-axis. Doing this using plt below.

```python
# Plot the scatter plot using plt
# Give title to plot and x axis label as age and y axis label as charges
plt.figure(figsize=(10,6))
plt.scatter(age, charges, alpha=0.5, color='blue')
plt.title('Scatter plot of Age vs Charges')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

Plotting scatter plot using seaborn below which is another way to plot scatter plots

```python
# Plot the scatter plot using sns  - another way of plotting scatter plots
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='charges')
plt.title('Scatter Plot of Age vs. Charges')
plt.show()
```

Using linear_model.LinearRegression() from sklearn, fit a model to
your data, and make predictions on the data

```python
# Perform linear regression on the scatter plot to find line of best fit

X = df[['age']] # Extracting age as feature variable
y= df[['charges']] # Extracting charges as target variable
model = LinearRegression() # Initialise a linear regression model
model.fit(X,y) # Train model on data and find line of best fit
predictions = model.predict(X); # Make predictions based on feature variable X
```

Plot another scatter plot with the best-fit line. Using plt to plot both scatter plot and best line of fit below.

```python
# Show line of best fit on scatter plot using plt

plt.figure(figsize=(10,6))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, predictions, color='red',label='Best-fit Line')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Scatter plot of Age vs Charges with Best-fit Line')
plt.legend()
plt.show()
```

Plotting scatter plot using seaborn and then plotting line of best fit 

```python
# Show line of best fit on scatter plot using sns and best-fit line using plt

plt.figure()
sns.scatterplot(data=df, x='age', y='charges', label='Actual Data')
plt.plot(X, predictions, color='red',label='Best-fit Line')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Scatter plot of Age vs Charges with Best-fit Line')
plt.legend()
plt.show()
plt.close()
```
